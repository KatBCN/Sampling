#### Reservoir Sampling

Reference:
- https://en.wikipedia.org/wiki/Reservoir_sampling

The following code block is taken from [The 5 Sampling Algorithms every Data Scientist need to know](https://towardsdatascience.com/the-5-sampling-algorithms-every-data-scientist-need-to-know-43c7bc11d17c)

```python
import random
def generator(max):
    number = 1
    while number < max:
        number += 1
        yield number
# Create as stream generator
stream = generator(10000)
# Doing Reservoir Sampling from the stream
k=5
reservoir = []
for i, element in enumerate(stream):
    if i+1<= k:
        reservoir.append(element)
    else:
        probability = k/(i+1)
        if random.random() < probability:
            # Select item in stream and remove one of the k items already selected
             reservoir[random.choice(range(0,k))] = element
print(reservoir)
#------------------------------------
#[1369, 4108, 9986, 828, 5589]
```

I slightly modify the code to be compatible for sampling from my data set.

```python
# Setting stream to number of rows in data frame to sample.
stream = generator(len(df))
# Doing Reservoir Sampling from the stream
#Setting k = to size of desired sample
k=100
reservoir = []
for i, element in enumerate(stream):
    if i+1<= k:
        reservoir.append(element)
    else:
        probability = k/(i+1)
        if random.random() < probability:
            # Select item in stream and remove one of the k items already selected
             reservoir[random.choice(range(0,k))] = element
```

```python
#reset the dataframe to the original data set.
df = pd.read_csv(data_link)

#create a new dataframe with using the index created by the reservoir sampling
reservoir_df = df.iloc[reservoir]
reservoir_df.head()
```

```python
reservoir_competition_counts = reservoir_df['Competition Title'].value_counts()
print (reservoir_competition_counts)

#Creating pie chart for distribution of number of records per competition
#define data
data = reservoir_competition_counts
labels = reservoir_competition_counts.index

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

We can see in the reservoir sample that the proportion of records from each competition has also changed slightly. The 2019 Youth World Championships increased from 68% to 71%, the 2018 Pan American Championships maintained its proportion, the 2019 World Championship decreased from 14% to 11%,  and the 2018 World Championships maintained its proportion.
